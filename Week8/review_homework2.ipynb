```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set()
```

```python
train = pd.read_csv('data/train.csv')
train
```

```python
maleSurv = train.loc[(train['Sex']=='male') &
                    (train['Age']>=40.0) &
                    (train['Survived']==1),'PassengerId'].count()
femaleSurv = train.loc[(train['Sex']=='female') &
                    (train['Age']>=40.0) &
                    (train['Survived']==1),'PassengerId'].count()
maleDie = train.loc[(train['Sex']=='male') &
                    (train['Age']>=40.0) &
                    (train['Survived']==0),'PassengerId'].count()
femaleDie = train.loc[(train['Sex']=='female') &
                    (train['Age']>=40.0) &
                    (train['Survived']==0),'PassengerId'].count()

print(maleSurv, femaleSurv, maleDie, femaleDie)

df = pd.DataFrame({
    'male': [maleDie, maleSurv],
    'female':[femaleDie,femaleSurv]
})
df
```
