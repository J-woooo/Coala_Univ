```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
import pandas as pd
train = pd.read_csv('data/train.csv')
train
```

```python
deadCount = train.loc[train['Survived']==0,'PassengerId'].count()
surviveCount = train.loc[train['Survived']==1,'PassengerId'].count()
print(deadCount)
print(surviveCount)

# 1ë‹¨ê³„
plt.bar([0,1],[deadCount,surviveCount])
plt.show()
```

```python
# 2ë‹¨ê³„
plt.bar(['0'],[deadCount],color = 'blue',width=0.5)
plt.bar(['1'],[surviveCount],color = 'red',width=0.5)
plt.ylabel('count')
plt.xlabel('Survived')
plt.legend(['Dead','Survived'])
plt.show()
```
