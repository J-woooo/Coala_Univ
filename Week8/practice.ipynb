```python
train = pd.read_csv('data/train.csv')
train
```

```python
import numpy as np
import seaborn as sns
sns.set()
```

```python
sns.relplot(data=train,x='Age',y='Survived')
```

```python
sns.relplot(data=train,x='Age',y='Survived',hue='Sex',aspect=3)
```

```python
sns.catplot(data=train,x='Pclass',y='Fare',aspect=3,hue='Sex')
```

```python
sns.pairplot(data=train,hue='Sex',
             x_vars=['Pclass','Age','Fare'],
             y_vars=['Survived','Age','Fare'],
             height=3)
```

```python
sns.catplot(data=train,x='Pclass',y='Fare',hue='Sex',kind='box')
```
