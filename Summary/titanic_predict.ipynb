### 데이터 불러오기

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data\\train.csv')
df.head(10)
```

### Feature Engineering

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
df.head(10)
print(df.isnull().sum())
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df.head()
```

```python
df['Cabin'] = df['Cabin'].fillna('C')
df['Cabin'] = df['Cabin'].str[0]
df.head()
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'] = df['Embarked'].fillna('S')
df.isnull().sum()
```

```python
df['Cabin'].value_counts()
```

```python
df['Cabin'] = df['Cabin'].map({
    'C':0,
    'B':1,
    'D':2,
    'E':3,
    'A':4,
    'F':5,
    'G':6,
    'T':7
})
df['Cabin'].value_counts()
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'] = df['Embarked'].map({
    'S':0,
    'C':1,
    'Q':2
})
df['Embarked'].value_counts()
```

```python
df['Sex'].value_counts()
```

```python
df['Sex'] = df['Sex'].map({
    'male':0,
    'female':1
})
df['Sex'].value_counts()
```

```python
train = df.drop(['Name','Ticket','SibSp','Parch','PassengerId'], axis=1)
test = df.drop(['Survived','Name','Ticket','SibSp','Parch','PassengerId'], axis=1)

train.head()
```

### IsAlone열을 추가하여  혼자탔는지 여부 파악

```python
df['IsAlone'] = 0
df.loc[df['FamilySize'] == 0,'IsAlone'] = 1
df
```

### Training Set과 Validation Set 나누기

```python
from sklearn.model_selection import train_test_split

train_data = train.drop('Survived',axis=1)
target_data = train['Survived']

x_train, x_valid, y_train, y_valid = train_test_split(train_data, target_data)

print(train_data.shape, x_train.shape, x_valid.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(x_train,y_train)

print("training set accuracy",tree.score(x_train,y_train))
print("valid set accuracy",tree.score(x_valid,y_valid))
```

### Test data를 predict하고 얻은 최종 결과를 csv 파일로 저장하기

```python
prediction = tree.predict(test)
prediction
```
