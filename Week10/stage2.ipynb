```python
import pandas as pd
df = pd.read_csv('house.csv')
df.head()
```

```python
df.describe()
```

```python
house_data = df.drop(['id', 'date'], axis = 1)

house_data
```

```python
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()
```

```python
plt.figure(figsize = (20,15))
sns.heatmap(house_data.corr(),annot=True,fmt='.2f',square=True)
plt.show()
```

```python
house_data.corr()
```

```python
sns.catplot(data=house_data, x='bathrooms', y='price',aspect=2)
```

```python
house_data['sqft_living'].max()
```

```python
minPrice = house_data['price'].min()
print(minPrice)
maxPrice = house_data['price'].max()
print(maxPrice)
averPrice = house_data['price'].mean()
print(averPrice)
```
